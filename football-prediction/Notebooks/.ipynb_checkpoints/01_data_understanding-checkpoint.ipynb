```python
import pandas as pd
from pathlib import Path
```

```python
data_path = Path("../data/raw")
files = list(data_path.glob("*.csv"))
```

```python
len(files)
```

```python
dfs = []

for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

data = pd.concat(dfs, ignore_index=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns.tolist()
```

```python
data = data[["Date", "HomeTeam", "AwayTeam", "FTHG", "FTAG", "FTR"]]
```

```python
data = data.rename(columns={
    "Date": "date",
    "HomeTeam": "home_team",
    "AwayTeam": "away_team",
    "FTHG": "home_goals",
    "FTAG": "away_goals",
    "FTR": "result"
})
```

```python
data.head()
```

```python
data.info()
```

```python
data.to_csv("../data/processed/matches.csv", index=False)
```
